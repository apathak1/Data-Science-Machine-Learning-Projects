```python
crash1 = crash.drop(columns=['TIME', 'FLIGHT#', 'ROUTE', 'AIRCRAFT_REG', 'CN_LN', 'TOTAL_ONBOARD', 'PASSENGER_ONBOARD', 'CREW_ONBOARD', 'GROUND_CASUALTIES', 'SUMMARY_OF_EVENTS'])
```

```python
#Checking to see how many null values, NA's, blank cells are there there in the data
crash1.isnull().sum()
```

```python
# removing NA's from the 'Location' column
crash2 = crash1[crash1['LOCATION'].notna()]
```

```python
# removing NA's from the 'Aircraft Type' column
crash3 = crash2[crash2['AIRCRAFT_TYPE'].notna()]
```

```python
# removing NA's from the 'Total Fatalities' column
crash4 = crash3[crash3['TOTAL_FATALITIES'].notna()]
```

```python
# removing NA's from the 'Total Fatalities' column
crash5 = crash4[crash4['PASSENGER_FATALITIES'].notna()]
```

```python
# removing NA's from the 'Crew Fatalities' column
crashclean = crash5[crash5['CREW_FATALITIES'].notna()]
```

```python
#Checking to see how many null values, NA's, blank cells are there there in the data
crashclean.isnull().sum()
```
