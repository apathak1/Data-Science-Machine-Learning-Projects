```python
import urllib, requests, urllib.parse, urllib.error
import json
import pandas as pd
from pandas import DataFrame
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Acquiring the data from API and checking status
response = requests.get('https://data.ontario.ca/api/3/action/datastore_search?resource_id=455fd63b-603d-4608-8216-7d8647f43350')
print(response)
```

```python
print(response.text)
```

```python
#API information in JSON format
print(response.json())
```

```python
# API information in an organzied format in JSON
content = json.dumps(response.json(), sort_keys = True, indent = 4)
print(content)
```

```python
# process of putting information into a dataframe and into a csv file
covid_list=[]

for lines in response.iter_lines(decode_unicode=True):
   if lines: 
      myline=json.loads(lines)
      print(mylines)
      covid_list.append(mylines)

# Pandas dataframe of the API information
mydf = pd.DataFrame(covid_list)

# CSV copy of the information
mydf.to_csv("covid_list.csv", header=True, index=False)
```

```python
print(mydf)
```

```python
#Cleaning process of the data
#Renaming the columns
mydf.rename(columns={"Test_Reported_Date":"Test Reported Date", "Age_Group":"Age Group", "Client_Gender":"Gender", "Case_AcquisitionInfo":"Information on Acquisition", "Outcome1":"Outcome", "Reporting_PHU":"Case Reporting Destination", "Reporting_PHU_City":"Case Reporting City", "Reporting_PHU_Latitude":"Case Reporting Latitude", "Reporting_PHU_Longitude":"Case Reporting Longitude"}, inplace=True)
print(mydf.columns)
```

```python
# Dropped these columns because I think they are unnecesary
covid_2 = mydf.drop(columns=['_id', 'Accurate_Episode_Date', 'Case_Reported_Date', 'Specimen_Date', 'Outbreak_Related', 'Reporting_PHU_Address', 'Reporting_PHU_Postal_Code', 'Reporting_PHU_Website'])
```

```python
print(covid_2)
```

```python
#Checking if the data has any N/A,s null values etc
covid_2.isnull().values.any()
```

```python
#Checking to see how many null values, NA's, blank cells are there there in the data
covid_2.isnull().sum()
```

```python
#removed N/A's and missing values in those two specific columns becuase N/A's are only present in those columns
covid_list_clean = covid_2.dropna(subset=['Test Reported Date', 'Gender'])
```

```python
# Re-Checking if the data has any N/A,s null values etc
covid_list_clean.isnull().values.any()
```

```python
#Re-Checking to see how many null values, NA's, blank cells are there there in the data
covid_list_clean.isnull().sum()
```
