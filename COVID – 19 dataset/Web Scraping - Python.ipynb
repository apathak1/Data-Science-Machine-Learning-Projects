```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==7:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
```

```python
#this gives the output into a dataframe for Table 1
import pandas as pd
df=pd.DataFrame(A,columns=['Public Health Unit'])
df['Cases']=B
df['Cases per m']=C
df['Resolved']=D
df['Death']=E
df['Deaths per m']=F
df
```

```python
#cleaning the website data - table 1
#removing the new lines symbol from data
df = df.replace('\n','', regex=True)
df
```

```python
#Renaming the columns
df.rename(columns={"Public Health Unit":"Town/Region", "Cases per m":"Cases/million inhabitants"}, inplace=True)
print(df.columns)
```

```python
#Checking if the data has any N/A,s null values etc
df.isnull().values.any()
```

```python
#Checking to see how many null values, NA's, blank cells are there there in the data
df.isnull().sum()
```

```python
#Checking if there are any duplicated values
df.duplicated()
```

```python
#Dropped this column
df_clean = df.drop(columns=['Deaths per m'])
```

```python
print(df_clean)
```
